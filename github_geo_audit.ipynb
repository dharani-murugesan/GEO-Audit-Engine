```python
# 1. Install and Configure the GenAI Stack
!pip install -U -q google-genai
import os
from google import genai
from google.colab import userdata

# Securely fetching the API key (Sanitized for GitHub)
# Instructions: Users should add their key to Colab Secrets as 'GOOGLE_API_KEY'
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
client = genai.Client(api_key=GOOGLE_API_KEY)

# Audit available models to ensure compatibility with 'generateContent'
models = client.models.list()
print("✅ SDK Setup Complete. Available Models for Audit:")
for m in models:
    if 'generateContent' in m.supported_actions:
        print(f"Model ID: {m.name}")
```

```python
import time

def run_geo_audit_safe(target_brand, industry_query, model_name='gemini-2.5-flash'):
    """
    Core diagnostic function with built-in Resilience Engineering.
    Handles 429 Rate Limit errors via an automated back-off loop.
    """
    combined_prompt = f"""
    Answer this query like a helpful search engine: "{industry_query}"
    Then, analyze your answer:
    1. Is "{target_brand}" mentioned? (Yes/No)
    2. Sentiment? (Positive/Neutral/Negative)
    3. Rank? (1 for first, 0 if not mentioned)

    Return format:
    ANSWER: [Your answer]
    MENTIONED: [Yes/No] | SENTIMENT: [Sentiment] | RANK: [Number]
    """

    # This loop manages API stability [cite: 162-175]
    for attempt in range(3):
        try:
            response = client.models.generate_content(model=model_name, contents=combined_prompt)
            return response.text
        except Exception as e:
            if "429" in str(e):
                print(f"⚠️ Rate limit hit. Waiting 30s... (Attempt {attempt+1}/3)")
                time.sleep(30)
            else:
                return f"❌ Error: {e}"
    return "❌ Failed after 3 attempts."
```

```python
from concurrent.futures import ThreadPoolExecutor

# Sanitized configuration for Portfolio visibility
TARGET_ENTITY = "Your_Target_Brand"
KEYWORDS = [
    "Top-rated service providers in [Sector]",
    "Leading [Industry] innovations 2026",
    "Best customer-rated [Category]"
]

print(f"🚀 Initiating Multi-Threaded Audit for: {TARGET_ENTITY}...")

# ThreadPoolExecutor reduces total runtime by running queries in parallel
with ThreadPoolExecutor(max_workers=5) as executor:
    results = list(executor.map(lambda k: run_geo_audit_safe(TARGET_ENTITY, k), KEYWORDS))

for i, res in enumerate(results):
    print(f"\n--- AUDIT KEYWORD: {KEYWORDS[i]} ---")
    print(res)
```

```python
# Demonstrating the use of the tool for a Deep Gap Analysis
TARGET_BRAND = "Target_Entity_Name"
INDUSTRY = "[Insert Industry]"

audit_query = f"""
Identify why '{TARGET_BRAND}' is missing from {INDUSTRY} results.
Provide a 4-step technical roadmap to improve GEO visibility,
focusing on Entity Association, Geographic Anchoring, and 'Key Considerations'.
"""

print(f"🔍 Running Strategic Roadmap Analysis for: {TARGET_BRAND}...")
report = run_geo_audit_safe(TARGET_BRAND, audit_query)
print(report)
```
